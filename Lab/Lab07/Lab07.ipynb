# Lab 10: Spectrogram

```python
%matplotlib inline

import sys
sys.path.append('../ThinkDsp')
import thinkdsp
import thinkplot
```

## Linear chirp

```python
signal1 = thinkdsp.Chirp(start=220, end=880)
wave1 = signal1.make_wave(duration=2)
```

```python
wave1.make_audio()
```

Notice that the pitch rises quickly at first and then slows down

```python
wave1.plot()
```

```python
wave1.segment(0,0.05).plot()
```

```python
wave1.segment(1,0.05).plot()
```

```python
wave1.segment(1.9,0.05).plot()
```

```python
sp1 = wave1.make_spectrum()
sp1.plot(high=1500)
```

Spectrum cannot show the relationship between frequency and time

```python
sg1 = wave1.make_spectrogram(seg_length=512)
sg1.plot(high=1000, shading='auto')
```
